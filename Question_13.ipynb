```python
#Import the required packages
from qiskit import QuantumCircuit
from qiskit.visualization import *
```

# Circuit depth 
Reference : https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.depth.html

NOTE: Quantum Circuit (Quantum Circuit Properties : Supplementary section)

Reference : https://qiskit.org/documentation/apidoc/circuit.html?highlight=quantum%20circuit%20depth

```python
qc=QuantumCircuit(2,2)

qc.h(0)

qc.barrier(0)
qc.cx(0,1)
qc.barrier([0,1])

qc.draw('mpl')
```

```python
qc.depth()
```

# IMP:
- Depth doesn't count the number of gates present in the qubit


# Let's take a 3-qubit example and see the results for the depth

```python
qc0=QuantumCircuit(3)
qc0.h(range(2))
qc0.x(2)
qc0.draw(output="mpl")
```

```python
qc0.depth()
```

```python
#Let's apply a barrier gate with no arguments
qc1=QuantumCircuit(1)
qc1.h(0)
qc1.h(0)
qc1.y(0)
qc1.draw(output="mpl")
```

```python
qc1.depth()
```
