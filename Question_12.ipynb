# Let's Simply try to write the code for the given circuit

```python
qc=QuantumCircuit(1,1)

qc.h(0)
qc.t(0)
qc.barrier()
qc.t(0)
qc.h(0)
qc.measure(0,0)

qc.draw('mpl')
```

```python
backend=Aer.get_backend('statevector_simulator')
job=execute(qc,backend)
result=job.result()
sv=result.get_statevector()
sv.draw(output="latex")
```
